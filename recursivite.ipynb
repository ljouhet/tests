```python
from rcviz import viz
```

```python
def trianglesB(n):
    if n > 0:
        trianglesB(n-1)
        print(n * "*")
```

```python
trianglesB(2)
```

```python
trianglesB(8)
```

```python
def factorielle(n: int) -> int:
    print(f"factorielle: n = {n}")
    if n == 0:
        print("factorielle(0) = 1")
        return 1
    else:
        resultat = n * factorielle(n-1)
        print(f"factorielle({n}) = {resultat}")
        return resultat
```

```python
factorielle(3)
```

```python
factorielle(10)
```

```python
@viz
def fibonacci(n : int) -> int:
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(4)
fibonacci.callgraph()
```

```python
fibonacci(6)
fibonacci.callgraph()
```
